## Read XML

```r
library(XML)
library(RCurl)
```

```r
url <- 'https://www.w3schools.com/xml/simple.xml'

xData <- getURL(url)
doc <- xmlTreeParse(xData, useInternal = TRUE)

root_node = xmlRoot(doc)
xmlName(root_node)
```

```r
names(root_node)
```

```r
root_node[[1]]
```

```r
root_node[[1]][[1]]
```

```r
xpathSApply(root_node, '//name', xmlValue)
```

```r
xpathSApply(root_node, '//price', xmlValue)
```

```r
# more complicate example
url <- 'https://www.espn.com/nfl/team/_/name/bal/baltimore-ravens'

xData <- getURL(url)
doc <- htmlTreeParse(xData, useInternal = TRUE)

root_node = xmlRoot(doc)
```

```r
xpathSApply(root_node, "//div[@class='score']", xmlValue)
```

```r
xpathSApply(root_node, "//div[@class='game-info']", xmlValue)
```

## Use data.table

Inherit from data.frame so all functions that work on data.fame work on data.table

Writen in C so it is much fater

```r
library(data.table)
```

```r
df = data.frame(x = rnorm(9), y = rep(c('a', 'b', 'c'), each = 3), z = rnorm(9))
df
```

```r
dt = data.table(x = rnorm(9), y = rep(c('a', 'b', 'c'), each = 3), z = rnorm(9))
dt
```

```r
# list all tables in the memory
tables()
```

```r
# subsetting rows
dt[2,]
```

```r
dt[dt$y == 'a',]
```

```r
dt[c(2, 3),]
```

```r
# subsetting column
dt[, c(2, 3)]
```

```r
# `expression` in R
k = {print(10); 5; 7}
```

```r
print(k)
```

```r
dt[, list(mean(x), sum(z))]
```

```r
dt[, table(y)]
```

```r
# adding new column
dt[, w:=x**2]
dt
```

```r
dt[, m:={tmp <- (x + z); log2(tmp + 10)}]
dt
```

```r
dt[, a:= x > 0]
dt
```

```r
dt[, b:=mean(x + w), by=a]
```

```r
dt
```

Special variables

.N an interger, length 1

```r
dt = data.table(x=sample(letters[1:3], 1E5, TRUE))
dt[, .N, by=x]
```

```r
# key
dt = data.table(x=rep(c('a', 'b', 'c'), each=100), y=rnorm(300))
setkey(dt, x)
dt['a']
```

```r
# join
dt1 = data.table(x=c('a', 'b', 'c', 'dt1'), y=1:4)
dt2 = data.table(x=c('a', 'a', 'c', 'dt2'), z=3:6)
setkey(dt1, x)
setkey(dt2, x)
merge(dt1, dt2)
```
